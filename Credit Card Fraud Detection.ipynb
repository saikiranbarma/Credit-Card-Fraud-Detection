## Feature Engineering
   #### Strandization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data['Amount'] = scaler.fit_transform(data[['Amount']])
data.head()
```

## Machine Learning Model
  #### Train_Test_Split

```python
from sklearn.model_selection import train_test_split
#splitting the data

x=data.drop('Class',axis=1)
y=data['Class']

# Split the data (stratify ensures proportional class distribution)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=42)

 #Check the shape of the splits
print(f"Training set size: {X_train.shape}, Test set size: {X_test.shape}")
```

#### Data Balancing

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Check the balanced distribution after SMOTE
print(f"Class distribution after SMOTE:\n{y_train_smote.value_counts()}")
```

#### Model Training and Evaluation
#### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
# Train a Logistic Regression model
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train_smote, y_train_smote)

# Make predictions on the test set
y_pred_logistic = logistic_model.predict(X_test)

# Predict probabilities (required for ROC curve)
y_prob_logistic = logistic_model.predict_proba(X_test)[:, 1]  # Get probability for the positive class (fraud cases

# Evaluate the model

from sklearn.metrics import classification_report
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_logistic))
auc_logistic = roc_auc_score(y_test, y_prob_logistic)
print(f"Logistic Regression AUC: {auc_logistic:.2f}")
```

#### Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Check if the sample data size is correct
sample_size = 10000  # A smaller subset of the data
print("Sampling the data...")

# Ensure you sample data correctly
X_train_smote_sample = X_train_smote.sample(n=sample_size, random_state=42)
y_train_smote_sample = y_train_smote.sample(n=sample_size, random_state=42)

# Create the Random Forest model
rf_model = RandomForestClassifier(random_state=42, n_estimators=50, n_jobs=-1)  # Reduce n_estimators if necessary

# Train the model with the sampled data
print("Training Random Forest model...")
rf_model.fit(X_train_smote_sample, y_train_smote_sample)

# Predict probabilities (required for ROC curve)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]  # Get probability for the positive class

# Make predictions
print("Making predictions...")
y_pred_rf = rf_model.predict(X_test)

# Print unique classes in y_test and y_pred_rf
print("Unique classes in y_test:", set(y_test))
print("Unique classes in y_pred_rf:", set(y_pred_rf))

# Output classification report
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
auc_rf = roc_auc_score(y_test, y_prob_rf)
print(f"Random Forest AUC: {auc_rf:.2f}")
```

#### XGB Classifier model

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report

# Train the XGBoost model without early stopping
xgb_model = XGBClassifier(random_state=42)
xgb_model.fit(X_train_smote_sample, y_train_smote_sample)

 # Predict probabilities (required for ROC curve)
y_prob_xgb = xgb_model.predict_proba(X_test)[:, 1]  # Get probability for the positive class


# Make predictions on the test set
y_pred_xgb = xgb_model.predict(X_test)

# Evaluate the model
print("XGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))
auc_xgb = roc_auc_score(y_test, y_prob_xgb)
print(f"XGBoost AUC: {auc_xgb:.2f}")

```

## Model Performance Comparison

```python
# Step 4: Plot ROC curves for each model to visually compare performance

fpr_logistic, tpr_logistic, _ = roc_curve(y_test, y_prob_logistic)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_prob_xgb)

plt.figure(figsize=(10, 6))
plt.plot(fpr_logistic, tpr_logistic, label=f'Logistic Regression (AUC = {auc_logistic:.2f})')
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {auc_rf:.2f})')
plt.plot(fpr_xgb, tpr_xgb, label=f'XGBoost (AUC = {auc_xgb:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for random guessing
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend(loc='lower right')
plt.show()
```
